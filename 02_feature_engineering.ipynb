## Feature Engineering

En esta fase se preparan los datos para su uso en modelos de Machine Learning.

Transformaciones realizadas:

- Separación de variables predictoras y variable objetivo.
- Identificación de variables categóricas.
- Transformar variables categóricas en variables numéricas.
- División del dataset en conjunto de entrenamiento y prueba utilizando muestreo estratificado.

El dataset resultante queda preparado para la fase de modelado.

```python
# Importacón de librerias

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Carga del dataset

df = pd.read_csv("data/df_clean.csv")
```

```python
#Comprobamos el dataset de manera general

df.info()
df.head()
```

```python
# Separamos la variable objetivo

x = df.drop("Churn", axis=1)
y = df["Churn"]
```

```python
# Encontramos las variables categoricas

cat_cols = x.select_dtypes(include="object").columns
cat_cols
```

```python
# Transformamos las variables categóricas en variables numéricas binarias

x = pd.get_dummies(x, drop_first=True)
```

```python
# Comprobamos el dataset transformado

x.head()
```

```python
# Separamos el dataset en datasets de train y test 

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
# Guardamos una copia de los datasets de train y test

x_train.to_csv("data/x_train.csv", index=False)
x_test.to_csv("data/x_test.csv", index=False)
y_train.to_csv("data/y_train.csv", index=False)
y_test.to_csv("data/y_test.csv", index=False)
```
